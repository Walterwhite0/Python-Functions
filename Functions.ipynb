1. What is the difference between a function and a method in Python?
  - Function:

     A function is a block of reusable code.
     
     It is not tied to any object or data type.
     
     You can use it anywhere in your code by calling its name.
     
     It can be built-in (like len(), print()) or user-defined (like def my_function()).

     Example:
```
  def greet():
      print("Hello from function!")
  greet()  
```

   - Method:

     A method is also a function, but it belongs to an object (like a string, list, dictionary, etc.).

     You can only use it with the object it is attached to.

     It is called using the dot notation like object.method().

     Example:
```
name = "shubham"
print(name.upper())  
```     

2.  Explain the concept of function arguments and parameters in Python.
  - In Python, parameters are the variables listed inside the function definition, while arguments are the actual values you pass to the function when calling it. Parameters act like placeholders that receive the values (arguments) during execution. For example, in def greet(name):, name is a parameter. When you call it with greet("Shubham"), "Shubham" is the argument. Python also supports different types of arguments like positional, keyword, default, and variable-length. Understanding the difference between parameters and arguments helps you write more flexible and reusable functions.

     Example:
```
def greet(name):  # 'name' is a parameter
    print("Hello", name)
greet("Shubham")  # "Shubham" is an argument
```

3. What are the different ways to define and call a function in Python?
  - In Python, functions are reusable blocks of code that perform a specific task. There are different ways to define and call a function. You can define a function using the def keyword followed by the function name and parentheses. Functions can have parameters (inputs) and can return a value using the return statement. You can also define default parameters in a function so that if no value is passed, it uses the default one. Using functions makes your code more organized, reusable, and easier to understand.

5. What are iterators in Python and how do they differ from iterables?
  -  In Python, an iterator is an object that allows you to traverse through all the elements of a collection (like lists, tuples, etc.) one at a time using the next() function. An iterable is any object in Python that can be looped over (like using a for loop).
But to actually go through the elements one by one, the iterable must be converted into an iterator using the iter() function.

     Difference:

     Iterable: An object you can loop through (like list, string, tuple).

     Iterator: An object that remembers its position and gives one value at a time using next().

     Example:
```
numbers = [1, 2, 3]       # This is an iterable
it = iter(numbers)        # Convert to iterator
print(next(it))           # Output: 1
print(next(it))           # Output: 2
print(next(it))           # Output: 3
```

6. Explain the concept of generators in Python and how they are defined.
  -  Generators in Python are special kinds of functions that let you produce a sequence of values one at a time instead of creating and storing the entire sequence in memory at once. This makes them very memory-efficient, especially when working with large data.

     You define a generator function just like a normal function using def, but instead of return, you use the yield keyword to send back a value each time. When you call a generator function, it doesn’t run the code right away but returns a generator object, which you can iterate over using next() or a for loop to get values one by one.

     Example:
```
def my_generator():
    yield 1
    yield 2
    yield 3
gen = my_generator()
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
```

7. What are the advantages of using generators over regular functions?
  -  Generators have a few cool advantages over regular functions. First, they save a lot of memory because they don’t create and store the entire output at once — they generate values one by one when needed. This is especially helpful when working with big data or long sequences.

     Second, generators make your code faster and more efficient because they don’t waste time or space creating huge lists or data structures upfront.

     Third, they allow you to write cleaner and simpler code when dealing with sequences, as you don’t have to manage indexes or create extra variables.

     So overall, generators help in saving memory, improving performance, and writing easy-to-understand code when you need to produce values step by step instead of all at once.  

     Example:
```
def my_generator():
    for i in range(3):
        yield i  
gen = my_generator()
for value in gen:
    print(value)
```

8. What is a lambda function in Python and when is it typically used?
  -  A lambda function in Python is like a small, quick function you can write without giving it a name. It’s useful when you need a simple function just for a short time, usually inside another function or when you want to keep your code short and clean. Instead of writing a full function with def, lambda lets you write it in one line.

     People usually use lambda functions for things like sorting, filtering, or applying a small operation to a list or data quickly, without making a separate full function.

     Example:
```
square = lambda x: x * x

```

9. Explain the purpose and usage of the `map()` function in Python.
  -  The map() function in Python is used to apply a specific function to every item in an iterable (like a list) and return a new iterable (map object) with the results. It’s handy when you want to transform or process all elements in a collection without writing loops explicitly.

     Basically, instead of writing a loop to apply the same operation on each item, you use map() to keep your code shorter and cleaner.

     Example:
```
def double(num):
    return num * 2
numbers = [1, 2, 3, 4]
result = map(double, numbers)
print(list(result))  # Output: [2, 4, 6, 8]
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  -  map() is used when you want to apply a function to each item in a list (or any iterable) and get back a new list with the changed values. It transforms every element.

     filter() is for when you want to pick only those items from a list that satisfy a certain condition. It filters out the elements that don’t match.

     reduce() is a bit different — it takes all the items in a list and combines them into a single value by applying a function repeatedly. For example, adding all numbers or multiplying them together.

     Example:
```
from functools import reduce
nums = [1, 2, 3, 4, 5]
# map: double each number
doubled = list(map(lambda x: x*2, nums))  # [2, 4, 6, 8, 10]
# filter: keep only even numbers
evens = list(filter(lambda x: x%2 == 0, nums))  # [2, 4]
# reduce: multiply all numbers
product = reduce(lambda x, y: x*y, nums)  # 120
print(doubled)
print(evens)
print(product)
```

````python
# 1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total
````

````python
# 2. Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(s):
    result = ""
    for char in s:
        result = char + result
    return result

text = input("Enter a string: ")

reversed_text = reverse_string(text)
print("Reversed string:", reversed_text)
````

````python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_list(numbers):
    result = []
    for num in numbers:
        result.append(num ** 2)
    return result

original_list = [1, 2, 3, 4, 5]
squared_list = square_list(original_list)

print("Original list:", original_list)
print("Squared list:", squared_list)
````

````python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")
````

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b

fib = FibonacciIterator(10)

for num in fib:
    print(num)
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power)
````

````python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # .strip() removes leading/trailing whitespace/newlines

for line in read_file_line_by_line("example.txt"):
    print(line)
````

````python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# List of tuples
data = [(1, 5), (3, 1), (2, 8), (4, 3)]

# Sort based on the second element using lambda
sorted_data = sorted(data, key=lambda x: x[1])

# Output the result
print(sorted_data)
````

````python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# List of temperatures in Celsius
celsius = [0, 10, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
````

````python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    result = ''.join(filter(lambda ch: ch not in vowels, input_str))
    return result

text = "Hello World!"
filtered_text = remove_vowels(text)
print("Original String:", text)
print("Without Vowels:", filtered_text)
````
